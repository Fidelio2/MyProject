1.数据获取

```python
from crawler import *

year = input("请输入要查询的年份：")

#不同的爬虫实现方法
#data = get_30_univs(year)
data = get_500_univs(year)
```

2.数据保存与数据处理

```python
from data_process import *

limit = int(input())
data = load_data(data,limit) # 保留前N项数据
OUTPUT_DIR = Path("output")
OUTPUT_DIR.mkdir(exist_ok=True)

# 分别设置path
json_path = OUTPUT_DIR / "university_ranking.json"
csv_path = OUTPUT_DIR / "university_ranking.csv"

save_json(data, json_path)
save_csv(data, csv_path)
save_data_files(data, OUTPUT_DIR)

print(f"JSON 文件已保存到: {json_path}")
print(f"CSV 文件已保存到: {csv_path}")
```

3.数据可视化

```python
from visualize import *

setup_chinese_font() # 设置中文字体

plot_score_bar(data, OUTPUT_DIR / "score_bar_top.png", year)
plot_score_line(data, OUTPUT_DIR / "score_line_rank.png", year)
plot_location_distribution(data, OUTPUT_DIR / "location_distribution.png", year)
plot_type_distribution(data, OUTPUT_DIR / "type_distribution.png", year)

print("可视化文件已生成：")
print(f"- {OUTPUT_DIR / 'score_bar_top.png'}")
print(f"- {OUTPUT_DIR / 'score_line_rank.png'}")
print(f"- {OUTPUT_DIR / 'location_distribution.png'}")
print(f"- {OUTPUT_DIR / 'type_distribution.png'}")
```
